```python
import boto3
from secrets import ACCESS_KEY_ID, SECRET_ACCESS_KEY
```

```python
s3 = boto3.resource(
    service_name='s3',
    region_name='us-east-1',  # name of the region your bucket is in
    aws_access_key_id=ACCESS_KEY_ID,
    aws_secret_access_key=SECRET_ACCESS_KEY
)
```

```python
for bucket in s3.buckets.all():
    print(bucket)
```

```python
s3.Bucket('moritz-demo-bucket').download_file(Key='titanic.csv', Filename='titanic.csv')
```
